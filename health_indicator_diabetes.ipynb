```python
import dask.dataframe as dd
df = dd.read_csv('datasets/Health_Indicators_Diabetes.csv')
df.head()
```

```python
df.dtypes
```

```python
corr_all = df.corr()
corr_all.compute()
```

```python
corr_all.sort_values('Diabetes_012').compute()
```

```python
corr_all['absDiabetes_012'] = corr_all['Diabetes_012'].abs()
corr_all_new = corr_all.sort_values('absDiabetes_012', ascending=True)
corr_all_new['absDiabetes_012'].compute().head(5)
```
